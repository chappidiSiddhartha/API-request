```python
import requests
import pandas as pd
from IPython.display import display
```

```python
def extract_from_api(country,n=-1):
     # initializing an error variable to 0

    err=0
    #print({country.lower().replace(' ','%20')})
    
    
    #  converting 'country' to lowercase and replacing spaces with %20 and sending the API request
    req = requests.get(f"http://universities.hipolabs.com/search?country={country.lower().replace(' ','%20')}") 
    
    # if status code != 200 error variable is assigned to 1
    if req.status_code!= 200:
        err = 1
    # Converting the request output in JSON format
    js = req.json() 

            

    try:    
        if err==1:
            # raises an exception in case of an error
            raise Exception()
        df = pd.DataFrame()

        for i in js[0].keys():
         
           # Converting/Structuring the JSON output into a data frame
            df[i] = [j[i] for j in js]
           # print(df[i])
    # Returning no data upon an error
    except:
        return "No Data Found" 
    else:
        if n!=-1:
            print("Number of rows:",len(df))
            return df.head(n) #returns top n records
        else:
            return df # returns the whole data frame if n=-1
```

```python
extract_from_api("United States",10)
```

```python
dataframe=extract_from_api("United States",-1)
dataframe.to_excel("univ_data.xlsx")
```
